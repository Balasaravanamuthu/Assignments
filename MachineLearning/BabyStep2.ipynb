```python
# Load the pandas library. It contains the functions related with Table. Create Alias name as pd for referring the library
import pandas as pd
```

```python
#Step1: Data Collection
# Use the pandas function read_csv to read the CSV (Comma Separated Values file) from the root folder. Assign the output to the variable df.
df = pd.read_csv("Salary_Data.csv")
```

```python
# Print the df to check all the input data exist in the variable or not.
df
```

```python
#Step2: Since there is no issue with data, data preprocessing hasn't done.
#Step3: Split the Input & Output data
#By using the column name get the input & output data and assign to the separate variables. Since it is Linear regression it has 1 input and 1 output column
#Input data. 
independent = df[["YearsExperience"]]
#Output data.
dependent = df[["Salary"]]
```

```python
#print the input variable to check the values are correct.
independent
```

```python
#print the output variable to check the values are correct.
dependent
```

```python
#Step4: Split the Train and Test data
#By using the sklearn packages, load the train_test_split function alone.
from sklearn.model_selection import train_test_split
```

```python
#passing the input and output variables to this function and get the split data (Train - X, Y and Test - X, Y)
X_train, X_test, y_train, y_test = train_test_split(independent, dependent)
```

```python
#print the X Train value
X_train
```

```python
#print the X Test value
X_test
```

```python
#print the Y Train value
y_train
```

```python
#print the Y Test value
y_test
```

```python
#Step5: Create Model
#Load the LinearRegression class from the sklearn.linear_model package
from sklearn.linear_model import LinearRegression
```

```python
#Create object to the class
regressor = LinearRegression()
```

```python
#Train the model using the X Train and Y Train data
#Output model will be generated.
regressor.fit(X_train,y_train)
```

```python
#Get the Predicted values based on the Test data 
y_pred = regressor.predict(X_test)
y_pred
```

```python
#Step6: Evaluate the metrics
#import the r2_score function from the sklearn.metrics package
from sklearn.metrics import r2_score
```

```python
#calculate the r2 metrics based on Y test and Y pred values.
#assign the metrics result to the variable
# metric value should between 0 and 1. model will perform good if the metric value is closure to 1.
result_score = r2_score(y_test,y_pred)
result_score
```

```python
#Calulate the weight
weight = regressor.coef_
weight
```

```python
#Calculate the bias
bias = regressor.intercept_
bias
```

```python
# Linear regression equation is y=mx+c, m is slope/weight and c is origin/intercept
#Step 7: Save the model
#Using Picle library save the model to the local
import pickle
#Create a file name and assign to the variable for save the model
filename = "linear_model.sav"
```

```python
#using the dump function save the created model to the root folder with the file name defined earlier using write binary.
#After execution we can see the file created in the root folder
pickle.dump(regressor, open(filename,"wb"))
```

```python
#The created model used to predict any salary based on the input YearsOfExperiance
#Step 8: Load the model
#Load the model by locating the saved file using read binary and assign to the variable for prediction
result_model = pickle.load(open(filename,"rb"))
```

```python
#Step 9: Predict using model
#Step 10: Call to Action
final_result = result_model.predict([[0]])
final_result
```
